```python
from lp_random_gen import random_lp_int
import os 
import shutil
from qiskit.optimization.converters import QuadraticProgramToQubo

DIR = 'TEST_DATA/'
getPath = lambda fn : DIR+fn
```

```python
conv = QuadraticProgramToQubo()
shutil.rmtree(DIR, ignore_errors=True)
os.makedirs(DIR)
```

```python
# create random binary Linear Programs
# we want up to 33 qubits since  ibmq_qasm_simulator has 32

for i in range(3,25):
    if i > 5 :
        c = int(i/2)
    elif i > 3 :
        c = i-2
    else:
        c=i
    lp_bin = random_lp_int(c, i, 0, 1 )
    model = lp_bin.createLP("test_"+str(i),1,-5,5,-2,2)
    model.write_to_lp_file(getPath(model.name))
    qubo = conv.convert(model)
    if qubo.get_num_vars() > 32 :
        print(i)
        break
    
```
