```python
from lp_random_gen import RandomLP
import os 
import shutil

DIR = 'TEST_DATA/DENSE'
getPath = lambda fn : DIR+fn
```

```python
shutil.rmtree(DIR, ignore_errors=True)
os.makedirs(DIR)
```

```python
# create random binary Linear Programs
# we want up to 33 qubits since  ibmq_qasm_simulator has 32

for i in range(3,25):
    if i > 5 :
        c = int(i/2)
    elif i > 3 :
        c = i-2
    else:
        c=i
    lp_bin = RandomLP.create_random_binary_prog("test_"+str(i),c,i)              
    lp_bin.write_to_lp_file(getPath(lp_bin.name))
    if lp_bin.complexity() > 32 :
        print(i)
        break
    
```
